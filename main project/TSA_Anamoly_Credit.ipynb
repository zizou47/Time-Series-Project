```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
date_range = pd.date_range(start='2020-01-01', end='2022-12-31', freq='D')
metric1_values = np.random.rand(len(date_range)) * 100

# Create a DataFrame
df_sample = pd.DataFrame({'date': date_range, 'metric1': metric1_values})
df_sample.set_index('date', inplace=True)

# Resampling the data to get daily, monthly, and yearly sums of metric1
df_daily = df_sample['metric1'].resample('D').sum()
df_monthly = df_sample['metric1'].resample('M').sum()
df_yearly = df_sample['metric1'].resample('Y').sum()

# Plotting daily metric1
plt.figure(figsize=(12, 6))
plt.plot(df_daily.index, df_daily, linestyle='-', color='blue', label='Daily Metric1')
plt.title('Time Series of Daily Metric1')
plt.xlabel('date')
plt.ylabel('Metric1')
plt.grid(False)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Plotting monthly metric1
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly, linestyle='-', color='green', label='Monthly metric7')
plt.title('Time Series of Monthly metric7')
plt.xlabel('date')
plt.ylabel('metric7')
plt.grid(False)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataset with a failure column
np.random.seed(0)
date_range = pd.date_range(start='2020-01-01', end='2022-12-31', freq='D')
metric1_values = np.random.rand(len(date_range)) * 100
failure_values = np.random.randint(2, size=len(date_range))

# Create a DataFrame
df_sample = pd.DataFrame({'date': date_range, 'metric7': metric1_values, 'failure': failure_values})
df_sample.set_index('date', inplace=True)

# Plotting daily metric1 with failure indication
plt.figure(figsize=(12, 6))
plt.plot(df_sample.index, df_sample['metric7'], linestyle='-', color='blue', label='Daily Metric1')
plt.scatter(df_sample[df_sample['failure'] == 1].index, df_sample[df_sample['failure'] == 1]['metric7'], color='orange', label='Failure (1)')
plt.scatter(df_sample[df_sample['failure'] == 0].index, df_sample[df_sample['failure'] == 0]['metric7'], color='blue', alpha=0.5, label='No Failure (0)')
plt.title('Time Series of Daily Metric1 with Failures')
plt.xlabel('date')
plt.ylabel('metric7')
plt.grid(False)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Resampling the data to get monthly and yearly sums of metric1
df_monthly = df_sample['metric7'].resample('M').sum()
df_yearly = df_sample['metric7'].resample('Y').sum()

# Plotting monthly metric1 with failure indication
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly, linestyle='-', color='green', label='Monthly metric7')
plt.scatter(df_sample[df_sample['failure'] == 1].resample('M').sum().index, df_sample[df_sample['failure'] == 1]['metric7'].resample('M').sum(), color='orange', label='Failure (1)')
plt.scatter(df_sample[df_sample['failure'] == 0].resample('M').sum().index, df_sample[df_sample['failure'] == 0]['metric7'].resample('M').sum(), color='green', alpha=0.5, label='No Failure (0)')
plt.title('Time Series of Monthly metric7 with Failures')
plt.xlabel('date')
plt.ylabel('metric7')
plt.grid(False)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```
